```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sys
sys.path.insert(0, "/Users/mbornhol/git/neural-profiling")
from utils import eval
```

```python
pre =  '/Users/mbornhol/git/neural-profiling/pre-trained/results/'
baseline_sphere = pd.read_csv(pre +'precision_baseline_top20.csv')
resnet = pd.read_csv(pre +'resnet_prec_sphere_select.csv')
efficientnet = pd.read_csv(pre +'res_prec_spherized.csv')
trained_813 = pd.read_csv('../runs/813/813_results.csv')
trained_811 = pd.read_csv('../runs/811/811_results.csv')
```

```python
# %matplotlib qt
plt.style.use({'figure.facecolor':'white'})
plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(figsize=(12,8))
plt.title("Average precision @k")
plt.plot(baseline_sphere["k"], baseline_sphere["precision"], marker="o", color="green", label = 'cytominer baseline')
plt.plot(efficientnet["k"], efficientnet["precision"], marker="o", color="blue", label = 'efficientnet pretrained')
plt.plot(trained_811["k"], trained_811["precision"], marker="o", color="orange", label = 'trained 10 epochs')
plt.plot(trained_813["k"], trained_813["precision"], marker="o", color="red", label = 'trained 24 epochs')
plt.ylabel('average precision')
plt.xlabel('k')
plt.legend(loc = 'upper right', shadow=True)
plt.show()
```

```python

plt.style.use({'figure.facecolor':'white'})
plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(figsize=(12,8))
plt.title("Average recall @k, top 20 MOAs")
plt.plot(baseline_sphere["k"], baseline_sphere["recall"], marker="o", color="green", label = 'cytominer baseline')
plt.plot(efficientnet["k"], efficientnet["recall"], marker="o", color="blue", label = 'efficientnet pretrained')
plt.plot(trained_811["k"], trained_811["recall"], marker="o", color="red", label = 'trained 10 epochs')
plt.plot(trained_813["k"], trained_813["recall"], marker="o", color="orange", label = 'trained 24 epochs')
plt.ylabel('recall')
plt.xlabel('k')
plt.legend(loc = 'upper left', shadow=True)
plt.show()
```
