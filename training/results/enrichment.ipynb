```python
pre =  '/Users/mbornhol/git/neural-profiling/pre-trained/results/'
efficientnet_sphered = pd.read_csv(pre + 'res_enrichment_spherized.csv')
resnet_sphered = pd.read_csv(pre + 'resnet_enr_sphere_select.csv')
baseline_spherized = pd.read_csv(pre + 'res_enr_spherized.csv')
baseline_mad_rob = pd.read_csv(pre + 'res_enr_mar.csv')
trained_813 = pd.read_csv('../runs/813/813_enrichment.csv')
trained_811 = pd.read_csv('../runs/811/811_enrichment.csv')
trained_806 = pd.read_csv('../runs/806_1/806_enrichment.csv')
trained_910 = pd.read_csv('../runs/910/910_enrichment.csv')
```
