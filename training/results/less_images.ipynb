```python
acc917 = pd.read_csv('accuracy/917.csv')
acc919 = pd.read_csv('accuracy/919.csv')

plt.style.use({'figure.facecolor':'white'})
plt.rcParams.update({'font.size': 19})
fig, ax = plt.subplots(figsize=(12,8))
plt.title("flat vs cosine LR")
plt.xlabel('step')

ax2 = ax.twinx()
ax.plot(acc917['loss'], 'g-', label = 'big sample')
ax2.plot(acc917['acc'], 'g-', label = 'acc' )
ax2.plot(acc917['val_acc'], 'g.', label = 'val_acc')

ax.plot(acc919['loss'], 'r-', label = 'small sample')
ax2.plot(acc919['acc'], 'r-', label = 'acc' )
ax2.plot(acc919['val_acc'], 'r.', label = 'val_acc')

ax.legend(loc = 'right', shadow=True)
ax2.legend(loc = 'upper right', shadow=True)
ax.set_ylabel('loss', color='b')
ax2.set_ylabel('acc', color='g')

plt.show()
```

```python

res917 = pd.read_csv('../runs/917/917_results.csv')
res919 = pd.read_csv('../runs/919/919_results.csv')
```

```python
pre =  '/Users/mbornhol/git/neural-profiling/pre-trained/results/'
efficientnet_sphered = pd.read_csv(pre + 'res_enrichment_spherized.csv')
plt.style.use({'figure.facecolor':'white'})
plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(figsize=(12,8))
plt.title("Folds of Enrichment in Top Connections")
plt.plot(100*res917["enrichment_percentile"], res917["ods_ratio"], marker="o", color="green", label = 'flat LR (917)')
plt.plot(100*res919["enrichment_percentile"], res919["ods_ratio"], marker="o", color="orange", label = 'Cosine LR (919)')
plt.plot(100*efficientnet_sphered["enrichment_percentile"], efficientnet_sphered["ods_ratio"], marker="o", color="red", label = 'efficientnet pretrained')
plt.ylabel('ods ratio')
plt.xlabel('percentile')
ax.invert_xaxis()
plt.legend(loc = 'upper right', shadow=True)
plt.show()
```

```python
efficientnet_sphered = pd.read_csv(pre + 'res_prec_spherized.csv')
plt.style.use({'figure.facecolor':'white'})
plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(figsize=(12,8))
plt.title("Average precision @k")
plt.plot(res917["k"], res917["precision"], marker="o", color="green", label = 'flat LR (917)')
plt.plot(res919["k"], res919["precision"], marker="o", color="orange", label = 'Cosine LR (919)')
plt.plot(efficientnet_sphered["k"], efficientnet_sphered["precision"], marker="o", color="red", label = 'efficientnet pretrained')
plt.ylabel('average precision')
plt.xlabel('k')
plt.legend(loc = 'upper right', shadow=True)
plt.show()
```

```python
plt.style.use({'figure.facecolor':'white'})
plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(figsize=(12,8))
plt.title("Average recall @k, top 20 MOAs")
plt.plot(res917["k"], res917["recall"], marker="o", color="green", label = 'flat LR (917)')
plt.plot(res919["k"], res919["recall"], marker="o", color="orange", label = 'Cosine LR (919)')
plt.plot(efficientnet_sphered["k"], efficientnet_sphered["recall"], marker="o", color="red", label = 'efficientnet pretrained')
plt.ylabel('recall')
plt.xlabel('k')
plt.legend(loc = 'upper left', shadow=True)
plt.show()
```
