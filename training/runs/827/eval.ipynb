```python
title = 'Precision@K and Recall@K: Training on Top50'
index1=pre_25_in
label_1="epoch 25, in-Top50"
index2=pre_25_out
label_2="epoch 25, out-Top50"
index3=pre_8_in
label_3="epoch 8, in-Top50"
index4=pre_8_out
label_4="epoch 8, out-Top50"
fontsize = 11
```

```python
plt.style.use({'figure.facecolor':'white'})
plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(figsize=(9,7))
plt.title(title)
ax2 = ax.twinx()
ax.set_xlabel("k")
ax.set_ylabel('Precision')
ax2.set_ylabel('Recall')

lns1 = ax.plot(index1["k"], index1["precision"], marker=".", color="r", label=f'{label_1}')
lns2 = ax2.plot(index1["k"], index1["recall"], marker=".", linestyle='dashed', color="r")
lns = lns1 + lns2

if str(index2) != "None":
    lns3 = ax.plot(index2["k"], index2["precision"], marker=".", color="g", label=f'{label_2}')
    lns4 = ax2.plot(index2["k"], index2["recall"], marker=".", linestyle='dashed', color="g")
    lns = lns1 + lns2 + lns3 + lns4

if str(index3) != "None":
    lns5 = ax.plot(index3["k"], index3["precision"], marker=".", color="b", label=f'{label_3}')
    lns6 = ax2.plot(index3["k"], index3["recall"], marker=".", linestyle='dashed', color="b")
    lns = lns1 + lns2 + lns3 + lns4 + lns5 + lns6


lns7 = ax.plot(index4["k"], index4["precision"], marker=".", color="y", label=f'{label_4}')
lns8 = ax2.plot(index4["k"], index4["recall"], marker=".", linestyle='dashed', color="y")
lns = lns1 + lns2 + lns3 + lns4 + lns5 + lns6 + lns7 + lns8


# lns = lns1+lns2+lns3+lns4+lns5+lns6
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc='right', prop={'size': fontsize}, shadow=True, bbox_to_anchor=(1,0.4))

plt.show()
```
